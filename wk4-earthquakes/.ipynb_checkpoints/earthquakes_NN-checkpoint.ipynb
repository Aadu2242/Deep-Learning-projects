```python
%matplotlib inline
# import dependancies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. Load Data

```python
raw = pd.read_csv('database.csv')
pd.options.display.max_columns = None # Display full columns
raw.head()
#data['Source'].value_counts()
```

```python
df = raw[['Date','Time','Latitude','Longitude','Magnitude']]
print(df.shape)
print(df.head())
print(df.info())
```

## 2. Data Pre-processing

### 2.1 Covert Date and Time values
Apprantly, it is improper to use current Date and Time values as inputs to the Neural Network. So we have to transform these values.
 - Date: seperate month and day, and generate 2 corresponding columns in the datafram
 - Time: extract hour from the value, and generate new column in the datafram
 - Finally, we will convert these new feaures (month, day and hour) into dummy variables. Dummy variables can better train our NN.

```python
dates = pd.to_datetime(df['Date'])
#data = pd.concat([data, [year,month,day]], axis=1)
pd.options.mode.chained_assignment = None  # default='warn'
#df.loc[:, 'year'] = dates.apply(lambda d: d.strftime('%Y'))
df.loc[:,'month'] = dates.apply(lambda d: d.strftime('%b'))
df.loc[:,'day'] = dates.apply(lambda d: d.strftime('%d'))
print(df.head())
#print(df.info())

# Plotting earthquake magnitude by years
#df1=data[['year','Magnitude']].groupby('year',as_index =False).mean()
#df1.plot(x='year',y='Magnitude')
```

```python
#df['Time'].str[0:2]
df.loc[:,'hour'] = df['Time'].apply(lambda x: x[0:2])
print(df.head())
#print(df.tail())
```

```python
data = df #keep the original data
```

```python
dummy_fields = ['month', 'day','hour']
for each in dummy_fields:
    dummies = pd.get_dummies(data[each], prefix=each, drop_first=False)
    data = pd.concat([data, dummies], axis=1)
```

### 2.2 Cleaning unnecessary features

```python
fields_to_drop = ['Date', 'month', 'day','hour','Time']
data = data.drop(fields_to_drop, axis=1)
print(data.shape)
data.head()
```

### 2.3 features scaling

```python
quant_features = ['Latitude', 'Longitude', 'Magnitude']
# Store scalings in a dictionary so we can convert back later
scaled_features = {}
for each in quant_features:
    mean, std = data[each].mean(), data[each].std()
    scaled_features[each] = [mean, std]
    data.loc[:, each] = (data[each] - mean)/std
```

```python
print(scaled_features)
data.head()
```

### 2.4 Split data 
 - split into train, test and validation datasets
 - seperate features and labels

```python
# split data train and test first
pos_train = int(data.shape[0] * 0.8)
train = data[:pos_train] # training 80%
test = data[pos_train:] # test 20%

print(train.shape,test.shape)
```

```python
# Separate the data into features and labels
target_fields = ['Magnitude']
features, targets = train.drop(target_fields, axis=1), train[target_fields]
test_features, test_targets = test.drop(target_fields, axis=1), test[target_fields]
```

```python
# Split training data into training and validation set
pos_val = int(train.shape[0] * 0.8)
train_features, train_targets = features[:pos_val], targets[:pos_val]
val_features, val_targets = features[pos_val:], targets[pos_val:]
```

```python
print(train_features.shape,train_targets.shape,val_features.shape,val_targets.shape)
```

## 3. Build the Neural Network

```python
class NeuralNetwork(object):
    def __init__(self, input_nodes, hidden_nodes, output_nodes, learning_rate):
        # Set number of nodes in input, hidden and output layers.
        self.input_nodes =  input_nodes
        self.hidden_nodes = hidden_nodes
        self.output_nodes = output_nodes

        # Initialize weights
        self.weights_input_to_hidden = np.random.normal(0.0, self.hidden_nodes**-0.5, 
                                       (self.hidden_nodes, self.input_nodes))

        self.weights_hidden_to_output = np.random.normal(0.0, self.output_nodes**-0.5, 
                                       (self.output_nodes, self.hidden_nodes))
        self.lr = learning_rate
        
        # Activation function is the sigmoid function
        self.activation_function = lambda x : 1 / (1 + np.exp(-x))    
    
    def train(self, inputs_list, targets_list):
        # Convert inputs list to 2d array
        inputs = np.array(inputs_list, ndmin=2).T
        targets = np.array(targets_list, ndmin=2).T
        
        #### Implement the forward pass here ####
        ### Forward pass ###
        # TODO: Hidden layer
        hidden_inputs = np.dot(self.weights_input_to_hidden, inputs) # signals into hidden layer
        hidden_outputs = self.activation_function(hidden_inputs)# signals from hidden layer
        
        # TODO: Output layer
        final_inputs = np.dot(self.weights_hidden_to_output,hidden_outputs) # signals into final output layer
        final_outputs = final_inputs # signals from final output layer
        
        #### Implement the backward pass here ####
        ### Backward pass ###
        
        # TODO: Output error  
        output_errors = targets - final_outputs  # Output layer error is the difference between desired target and actual output.
        #output_grad = output_errors # y=x derivative is 1 -- 1x128
        
        # TODO: Backpropagated error
        hidden_errors = np.dot(self.weights_hidden_to_output.T,output_errors) # errors propagated to the hidden layer 2x128
        hidden_grad = hidden_outputs*(1-hidden_outputs)  # hidden layer gradients -- 2x128
        
        # TODO: Update the weights
        self.weights_hidden_to_output += self.lr*np.dot(output_errors,hidden_outputs.T) # update hidden-to-output weights with gradient descent step
        self.weights_input_to_hidden += self.lr*np.dot(hidden_errors*hidden_grad,inputs.T) # update input-to-hidden weights with gradient descent step
 
        
    def run(self, inputs_list):
        # Run a forward pass through the network
        inputs = np.array(inputs_list, ndmin=2).T
        
        #### Implement the forward pass here ####
        # TODO: Hidden layer
        hidden_inputs = np.dot(self.weights_input_to_hidden, inputs) # signals into hidden layer
        hidden_outputs = self.activation_function(hidden_inputs) # signals from hidden layer
        
        # TODO: Output layer
        final_inputs = np.dot(self.weights_hidden_to_output,hidden_outputs) # signals into final output layer
        final_outputs = final_inputs # signals from final output layer 
        
        return final_outputs
```

```python
# Mean Square Error cost function
def MSE(y, Y):
    return np.mean((y-Y)**2)
```

## 4. Train the Neural Network

```python
def randomSearch(epochs_batch, learning_rate_batch,hidden_nodes_batch,N_i,output_nodes):
    hyper_param =[]
    for hidden_nodes in hidden_nodes_batch:
        for epochs in epochs_batch:
            for learning_rate in learning_rate_batch:
                network = NeuralNetwork(N_i, hidden_nodes, output_nodes, learning_rate)
                for e in range(epochs):
                    batch = np.random.choice(train_features.index, size=128)
                    for record, target in zip(train_features.ix[batch].values,train_targets.ix[batch]['Magnitude']):
                        network.train(record, target)
                    train_loss = MSE(network.run(train_features), train_targets['Magnitude'].values)
                    val_loss = MSE(network.run(val_features), val_targets['Magnitude'].values)
                    
                hyper_param.append([(hidden_nodes,epochs,learning_rate),(train_loss,val_loss)]) 
                
    return hyper_param                  
                
```

```python
import sys

### Set a batch hyperparameters for each one ###
epochs_batch = [500,700,1000]
learning_rate_batch = [0.01,0.001]
hidden_nodes_batch = [10,22,33]
output_nodes = 1
N_i = train_features.shape[1]

hyperparameter_list = randomSearch(epochs_batch, learning_rate_batch,hidden_nodes_batch,N_i,output_nodes)
```

```python
for hyper in hyperparameter_list:
    print(hyper[0],"\t-->",hyper[1])
```

```python
minVal = 1.0
hyper_param = (0,0,0)
for hyper in hyperparameter_list:
    for h in hyper:
        if h[1] < minVal:
            minVal = h[1]
            hyper_param = hyper
            
print(hyper_param,"\t-->",minVal)
```

```python
import sys

### Set a hyperparameters from above results ###
epochs = 500
learning_rate = 0.01
hidden_nodes = 10
output_nodes = 1

N_i = train_features.shape[1]
network = NeuralNetwork(N_i, hidden_nodes, output_nodes, learning_rate)

losses = {'train':[], 'validation':[]}
for e in range(epochs):
    # Go through a random batch of 128 records from the training data set
    batch = np.random.choice(train_features.index, size=128)
    for record, target in zip(train_features.ix[batch].values, 
                              train_targets.ix[batch]['Magnitude']):
        network.train(record, target)
    
    # Printing out the training progress
    train_loss = MSE(network.run(train_features), train_targets['Magnitude'].values)
    val_loss = MSE(network.run(val_features), val_targets['Magnitude'].values)
    sys.stdout.write("\rProgress: " + str(100 * e/float(epochs))[:4] \
                     + "% ... Training loss: " + str(train_loss)[:5] \
                     + " ... Validation loss: " + str(val_loss)[:5])
    
    losses['train'].append(train_loss)
    losses['validation'].append(val_loss)
```

```python
plt.plot(losses['train'], label='Training loss')
plt.plot(losses['validation'], label='Validation loss')
plt.legend()
#plt.ylim(ymax=0.5)
```
